```python
ds=df.groupby('Location')
```

```python
datapts=[x[1] for x in ds]
```

```python
datapts[0]
```

```python
new_data=[]
for doc in datapts:
    new=[]
    raint=list(doc["RainTomorrow"].values)
    new_doc=doc[0:len(doc)-4]
    print(len(new_doc))
    for y in range(5):
        new=[]
        for x in range(len(raint)-y):
            if sum(raint[x:x+y+1])>0:
                new.append(1)
            else:
                new.append(0)
        new_doc.loc[:,"rain_"+str(y+1)+"_days"]=new[0:len(new)-(4-y)]
    new_data.append(new_doc)
            
    
```

```python
X.loc[X["RainToday"]=="Yes","RainToday"]=1
X.loc[X["RainToday"]=="No","RainToday"]=0
```

```python
X=X[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation',
       'Sunshine', 'WindGustSpeed',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RainToday', 'RISK_MM',"rain_2_days"]]
```

```python
X.columns
```

```python
X=X.dropna()
y=X["rain_2_days"]
del X["rain_2_days"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
t0=time.time()
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
clf_rf = RandomForestClassifier(n_estimators=100, max_depth=10,random_state=0)
clf_rf.fit(X_train,y_train)
y_pred = clf_rf.predict(X_test)
score = accuracy_score(y_test,y_pred)
print('Accuracy :',score)
print('Time taken :' , time.time()-t0)
```

```python
from sklearn.linear_model import LogisticRegression
t0=time.time()
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
clf_logreg = LogisticRegression(random_state=0)
clf_logreg.fit(X_train,y_train)
y_pred = clf_logreg.predict(X_test)
score = accuracy_score(y_test,y_pred)
print('Accuracy :',score)
print('Time taken :' , time.time()-t0)
```
